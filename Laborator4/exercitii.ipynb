```python
#Exercitiul 1
import pandas as pd

dataset = pd.read_csv("data.csv")
(dataset.head())
```

```python
#Exercitiul 2
rezultat = dataset[dataset['Age'] > 40]
rezultat.head(10)
```

```python
#Exercitiul 3
rezultat = dataset[(dataset['Overall'] >= 85) & (dataset['Age'] < 25)]
rezultat
```

```python
#Exercitiul 4
rezultat = dataset.sort_values(by='Skill Moves', ascending=False)
rezultat
```

```python
#Exercitiul 5
dataset.columns = dataset.columns.str.strip()

rezultat = dataset[dataset['Contract Valid Until'].astype(str).str.contains('2021')]
rezultat
```

```python
#Exercitiul 6
rows, cols = dataset.shape
print("Numar randuri:", rows)
print("Numar coloane:", cols)

unique_players = dataset["Name"].nunique()
print("Numar jucatori unici:", unique_players)
```

```python
#Exercitiul 7
top_5_nationalitati = dataset["Nationality"].value_counts().head(5)

print("Top 5 nationalitati ale jucatorilor sunt:")
print(top_5_nationalitati)

cea_mai_frecventa = dataset["Nationality"].value_counts().idxmax()

print("\nCea mai frecventa nationalitate este:")
print(cea_mai_frecventa)
```

```python
#Exercitiul 8
import matplotlib.pyplot as plt

top_5 = dataset["Nationality"].value_counts().head(5)

print("Top 5 nationalitati:")
print(top_5)

plt.figure(figsize=(7,7))
top_5.plot(kind="pie", autopct="%1.1f%%")

plt.title("Top 5 nationalitati")
plt.ylabel("")
plt.show()
```

```python
#Exercitiul 9
avg_stats = dataset.groupby("Nationality")[["SprintSpeed", "Acceleration"]].mean()

print("Media SprintSpeed si Acceleration pentru fiecare nationalitate:")
display(avg_stats)
```

```python
#Exercitiul 10

dataset["Position"] = dataset["Position"].fillna("Unknown")
```

```python
#Exercitiul 11

club_avg = dataset.groupby("Club")["Overall"].mean()

best_club = club_avg.sort_values(ascending=False).head(1)

print("Clubul cu cea mai mare medie de Overall este:")
print(best_club)
```

```python
def convert_money(x):
    if isinstance(x, str):
        x = x.replace('€', '').replace('K', '000').replace('M', '000000')
        return float(x)
    return x

dataset["Value"] = dataset["Value"].apply(convert_money)
dataset["Wage"] = dataset["Wage"].apply(convert_money)

count = (dataset["Value"] > dataset["Wage"]).sum()

print("Numar jucatori cu valoarea de transfer mai mare ca salariul:", count)
```

```python
#Exercitiul 13

dataset["is_underpaid"] = dataset["Wage"] < (dataset["Value"] / 100)

dataset[["Name", "Wage", "Value", "is_underpaid"]].head()
```

```python
#Exercitiul 14
dataset["Score"] = (
    0.3 * dataset["Overall"] +
    0.3 * dataset["Potential"] +
    0.2 * dataset["SprintSpeed"] +
    0.2 * dataset["ShortPassing"]
)

print("Scorul a fost calculat!")
display(dataset[["Name", "Score"]].head())
```

```python
#Exercitiul 15

good_deals = dataset[["Name", "Wage", "Value"]].copy()

good_deals["difference"] = good_deals["Value"] - good_deals["Wage"]

good_deals = good_deals.sort_values(by="difference", ascending=False)

print("Top jucatori care reprezinta afaceri bune:")
good_deals
```

```python
#Exercitiul 16

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.scatterplot(data=dataset, x="Wage", y="Value")

plt.title("Wage vs Value")
plt.show()
```

```python
#Exercitiul 17

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool

source = ColumnDataSource(dataset)

p = figure(title="Wage vs Value")

p.scatter(x="Wage", y="Value", source=source, size=8)

hover = HoverTool(tooltips=[("Name", "@Name"),
                            ("Value", "@Value"),
                            ("Wage", "@Wage")])

p.add_tools(hover)

show(p)
```
